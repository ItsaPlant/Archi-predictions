```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data (RESULTS)
df = pd.read_csv("C:\_PRYWATNE\LANGUAGES\Python\_DATA_SCIENCE\py\_ARCHITECTURE\PMV_PREDICTIONS\DATA\data.csv")
df['date'] = pd.to_datetime(df['date'], format='%d %b %H:%M').dt.strftime('%d %b %H:%M')
df.head()
```

```python
nan_df = df.isna()

# Plot the heatmap
plt.figure(figsize=(8, 6))  # You can adjust the size
sns.heatmap(nan_df, cmap="coolwarm", cbar=False, yticklabels=False)

# Show the heatmap
plt.title('NaN Heatmap')
plt.show()
```

```python
# Plotting a line chart for all variables
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot each column as a separate line
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

# Adding labels and title
plt.title("Line Chart for 8 Variables", fontsize=16)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Show legend
plt.legend(title="Variables")

# Show the plot
plt.show()
```

```python
df = df

# Set up the figure
plt.figure(figsize=(8, 6))  # Adjust figure size if needed

# Create the heatmap with vmin and vmax to control the color scaling
sns.heatmap(df, annot=False, cmap="YlGnBu", linewidths=0.5, annot_kws={"size": 12}, fmt='.1f', vmin=df.min().min(), vmax=df.max().max())

# Rotate y-axis labels if they overlap (optional)
plt.yticks(rotation=0)  # You can set this to rotate labels if needed
plt.xticks(rotation=45)  # Rotate x-axis labels if needed

# Apply tight layout to prevent annotation stacking and ensure the elements fit well
plt.tight_layout()

# Add a title
plt.title("Heatmap of DataFrame", fontsize=16)

# Show the heatmap
plt.show()
```
